```python
import pandas as pd
import numpy as np

indicator_dfs = []
for ind in ['总市值', '市盈率(TTM)', '市净率']:
    df = ak.stock_zh_valuation_baidu(symbol='603259', indicator=ind, period='全部')
    df = df.set_index('date').rename(columns={'value':ind})
    indicator_dfs.append(df)
df = pd.concat(indicator_dfs, axis=1)
print(df.sample(5))
```

## 双坐标方式一

```python
import akshare as ak
import pandas as pd
from matplotlib.figure import Figure
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(df.index, df['总市值'], 'g-')
ax2.plot(df.index, df['市盈率(TTM)'], 'b-')

ax1.set_xlabel('X data')
ax1.set_ylabel('Y1 data', color='g')
ax2.set_ylabel('Y2 data', color='b')

plt.show()
```

## 双坐标方式二

```python
from matplotlib.figure import Figure
fig = Figure(figsize=(8, 4), dpi=100)

#fig = plt.figure(figsize=(7, 4)) # 设置图形大小
ax1 = fig.subplots()

plt.plot(df['总市值'], 'b', lw=1.5, label='1ab')
plt.legend(loc=8)

ax2 = ax1.twinx()
plt.plot(df['市盈率(TTM)'], 'g', lw=1.5, label='2ab')
plt.legend(loc=0)
plt.ylabel('value 2nd')

plt.grid(True)
plt.axis('tight')
```
